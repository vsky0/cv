# ColorSpaces

there are different types of color spaces like, GRAY, HSV, BGR, RGB and more.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Helper function

this function converts the image to rbg.

```python
def display_img_in_rgb(image, title='image'):
    rgb_img = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.title(title)
    plt.axis('off')
    plt.show()
    
```

```python
img = cv.imread('../data/photos/cell100.jpg')
display_img_in_rgb(img, title="original image")
```

## BGR2GRAY

```python
gray_img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
display_img_in_rgb(gray_img, 'gray image')
```

## to HSV

HSV is Hue saturation value

it is how the we humans percive color.

```python
hsv_img = cv.cvtColor(img, cv.COLOR_RGB2HSV)
display_img_in_rgb(hsv_img, 'hsv image')
```

## LAB



```python
lab_img = cv.cvtColor(img, cv.COLOR_RGB2LAB)
display_img_in_rgb(lab_img, 'lab image')
```

## HSV2RGB

```python
hsv_bgr = cv.cvtColor(hsv_img, cv.COLOR_HSV2RGB)

fig,ax = plt.subplots(1,2)

ax[0] = display_img_in_rgb(hsv_img, 'hsv image')

ax[1] = display_img_in_rgb(hsv_bgr, 'hsv 2 rgb')
```

```python
lab_rgb = cv.cvtColor(lab_img, cv.COLOR_LAB2RGB)

fig,ax = plt.subplots(1,2)

ax[0].imshow(lab_img)
ax[0].set_title('LAB image')
ax[1] = display_img_in_rgb(lab_rgb, 'LAB2RGB')

plt.show()
```
